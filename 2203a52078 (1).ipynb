```python
import numpy as np

# Define lists as datasets
list1 = [2, 4, 6, 8, 10]
list2 = [1, 2, 3, 4, 5]

# Convert lists to NumPy arrays
array1 = np.array(list1)
array2 = np.array(list2)

# Perform multiplication element-wise using NumPy arrays
result_multiplication = array1 * array2
print("Multiplication Result (NumPy array):", result_multiplication)

# Perform division element-wise using NumPy arrays
result_division = array1 / array2
print("Division Result (NumPy array):", result_division)

# Perform exponentiation element-wise using NumPy arrays
result_poweroff = np.power(array1, array2)
print("Poweroff Result (NumPy array):", result_poweroff)

# If you want the results as lists, you can convert the arrays back to lists
result_multiplication_list = result_multiplication.tolist()
result_division_list = result_division.tolist()
result_poweroff_list = result_poweroff.tolist()

print("Multiplication Result (list):", result_multiplication_list)
print("Division Result (list):", result_division_list)
print("Poweroff Result (list):", result_poweroff_list)
```

```python
import numpy as np
list=[1, 2, 3, 4, 5]
# Create a NumPy array
array = np.array(list)

# Apply NumPy functions on the array
sin_result = np.sin(array)
log_result = np.log(array)
log2_result = np.log2(array)
exp_result = np.exp(array)

print("Original Array:")
print(array)

print("\nResult of np.sin():")
print(sin_result)

print("\nResult of np.log():")
print(log_result)

print("\nResult of np.log2():")
print(log2_result)

print("\nResult of np.exp():")
print(exp_result)
```

```python
import pandas as pd

import pandas as pd

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
health_condition = ['Good', 'Average', 'Good', 'Poor']

healthcare_data = {
    'People': people,
    'Age': age,
    'Height': height,
    'Weight': weight,
    'Health Condition': health_condition
}

healthcare_df = pd.DataFrame(healthcare_data)
print(healthcare_df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(healthcare_df['Age'], healthcare_df['Weight'])
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Scatter Plot: Age vs. Weight')
plt.show()
```

```python
import seaborn as sns

sns.countplot(x='Health Condition', data=healthcare_df)
plt.xlabel('Health Condition')
plt.ylabel('Count')
plt.title('Bar Chart: Health Condition Distribution')
plt.show()
```

```python
plt.hist(healthcare_df['Height'], bins=10, edgecolor='black')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Histogram: Height Distribution')
plt.show()
```

```python
plt.scatter(healthcare_df['Age'], healthcare_df['Height'])
plt.title('Scatter Plot: Age vs. Height')
plt.xlabel('Age')
plt.ylabel('Height (in cm)')
plt.grid(True)
plt.show()
plt.bar(healthcare_df['People'], healthcare_df['Weight'])
plt.title('Bar Chart: Person vs. Weight')
plt.xlabel('Person')
plt.ylabel('Weight (in kg)')
plt.grid(axis='y')
plt.show()
plt.hist(healthcare_df['Age'], bins=10, color='blue', edgecolor='black')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
